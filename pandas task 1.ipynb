```python
#importing libraries
import pandas as pd
import requests
```

```python
#setting url link
DATA_URL='https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
#importing raw content from url with requests.get
try:
    rawdata=requests.get(DATA_URL)
except Exception as ex:
    print(ex)
```

```python
type(rawdata)
#data extracted is Response type
```

```python
#convering rawdata into json
json_content=rawdata.json()
type(json_content)
```

```python
#making json into dataframe with the original columns 
main_df=pd.DataFrame(json_content, columns=['id','url','repository_url','labels_url','user'])
```

```python
#checking the df
main_df
```

```python
#need to seperate the items in 'user' by writing them in a new list
user_field_list=[]
for i in main_df['user']:
    user_field_list.append(i)
```

```python
user_field_list
#each entry is a dictionary and is added to a list
```

```python
#making a dataframe from the list
user_df=pd.DataFrame(user_field_list)
```

```python
user_df
```

```python
#combining the two dataframes: main and user using concat
final_df=pd.concat([main_df,user_df],axis=1,join='inner')
```

```python
final_df.drop(['user'],axis=1)
```

```python
#writing the result in a csv
final_df.to_csv('task1.csv')
```
